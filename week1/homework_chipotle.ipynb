```python
import csv
with open("data/order.tsv", mode="rU") as f:
    file_nested_list = [row for row in csv.reader(f, delimiter="\t")]
print(file_nested_list[0])
file_nested_list[1]    
```

```python
# separate header and data
header = file_nested_list[0]
data = file_nested_list[1:]
```

```python
data[0:5]
```

```python
# Getting all orders
all_orders = len([row[0] for row in data])
print(all_orders)

# getting unique orders
orders = len(set([row[0] for row in data]))
print (orders)
```

```python
row[-1][1:]
```

```python
#getting list of prices, ignoring the $ sign
prices = [float(row[-1][1:]) for row in data]
print(prices)
```

```python
# compute the average price
round(sum(prices)/orders, 2)

round(sum([float(row[-1][1:]) for row in data])/len(set([row[0] for row in data])),2)

```

```python
# another approach to remove $ sign in pricing using regular expression(regex)
import re
order_item_price = [float(re.sub(r'[^\d.]', '', row[4])) for row in data]
order_item_price
```

```python
#  get the list of sodas orders
sodas = []
for row in data:
    if "Canned" in row[2]:
        sodas.append(row[3])
sodas       
```

```python
#  list of burritos orders
burritos = len([row[2] for row in data if "Burrito" in row[2]])
burritos
```

```python
# list of toppings
toppings = [row[3].count(",")+1 for row in data if "Burrito" in row[2]]
toppings
```

```python
#  compute the average number of topping for burritos orders
import numpy as np
round(np.mean(toppings), 2)
#uses numpy
round(float(sum(toppings))/float(burritos),2)
```

```python
# Create a dictionary in which the keys represent chip orders and
# the values represent the total number of orders.
from collections import Counter
chips = Counter([row[2] for row in data if "Chips" in row[2]])
print chips
```

```python
idx_order_item = 2
idx_order_qty = 1

from collections import defaultdict
chips = defaultdict(int)
for row in data:
    if 'chips' in row[idx_order_item].lower():
        chips[row[idx_order_item].lower()] += int(row[idx_order_qty])
for key, value in sorted(chips.iteritems()):
    print key.rjust(48), ":", str(value).rjust(4)
    
print(chips.items())
```
